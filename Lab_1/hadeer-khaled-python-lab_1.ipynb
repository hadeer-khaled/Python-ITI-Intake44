## 2- Write a two functions to calculate the area and the perimeter of a rectangle given its length and width.


```python
def rectangle_perimeter(length , width):
    return 2 * (length + width)
    
def rectangle_area(length , width):
    return length * width

print(f"Rectangle Perimeter: {rectangle_perimeter(10 , 5)}")
print(f"Rectangle Area: {rectangle_area(10 , 5)}")
```

## 5- Implement a program to merge two dictionaries and return the result

```python
def dict_merge_inplace(dict1, dict2):
    for i in dict2.keys():
        dict1[i]=dict2[i]
    return dict1
        
def dict_merge(dict1 , dict2):
    return {**dict1, **dict2}

dict1 ={'key1': 1, 'key2': 2}
dict2 = {'key3': 3, 'key4': 4}

print("Method 1:", dict_merge_inplace(dict1,dict2 ))
print("Method 2:", dict_merge({'key1': 1, 'key2': 2}, {'key3': 3, 'key4': 4}))
```

## 12- Write a function that removes even numbers from a list and squares the remaining odd numbers.

```python
def sqr_odd(input_list):
    squared_odd_numbers = []
    for num in input_list:
        if num % 2 != 0:
            squared_odd_numbers.append(num ** 2)

    return squared_odd_numbers

print(sqr_odd([1,2,3,4,5,6,7]))
```

## 13- Implement a function to check if a number is prime or not.


```python
def prime_checker(num):
    for i in range(2, int(num/2) + 1):
        if num % i == 0:
            return False
    return True

num = 4;
print(f"Is {num} a prime number ?", prime_checker(num))
```

## 14- Write a function to calculate the factorial of a given number.


```python
def factorial(num):
    if num < 0:
        return "Cannot calculate the factorial for negative numbers"
    elif num == 0 or num == 1:
        return 1
    else:
        res = 1
        for i in range(2, num + 1):
            res *= i
        return res
num = 5  
print(f"The factorial of {num} = ", factorial(num))
```

1. Double Asterisk (**)

* Function Keyword Argument Unpacking: In function definitions or calls, ** is used to unpack key-value pairs from a dictionary and pass them as keyword arguments.
* Dictionary Unpacking: In dictionary unpacking, is used to unpack key-value pairs from one dictionary and include them in another.

2. Single Asterisk (*):
* Function Argument Unpacking: In function definitions or calls, * is used to unpack the elements of an iterable (e.g., a list, tuple) and pass them as separate arguments.
* Iterable Unpacking in Iterables: In iterable unpacking, * is used to unpack elements from one iterable and include them in another.
