## Question1

*     Define a class named Shape with attributes name and _color (private).

*     Implement encapsulation by providing methods get_color() and set_color(color).

*     Include an abstract method calculate_area() in the Shape class.
*     Create two subclasses, Circle and Square, each inheriting from Shape.
*     Implement polymorphism by overriding the calculate_area() method in the subclasses.
*     Create another subclass, Rectangle, inheriting from Shape with additional attributes such as length and width.
*     print the calculate_area for Circle, Square and Rectangle "assume length = width = 5 as inputs"

```python
from abc import ABC ,abstractmethod
from math import pi 

class Shape:
    def __init__(self , name , color):
        self.name = name;
        self._color = color;
        
    def set_color(self,color):
        self.color = color ;
        
    def get_color(self):
        return self.color
    
    @abstractmethod
    def calculate_area(self):
        pass
    
class Square(Shape):
    def __init__(self , name , color , side_length):
        super().__init__(name, color)
        self.side_length=side_length
        
    def  calculate_area(self):
        return round((self.side_length*self.side_length) ,2)
    
class Circle(Shape):
    def __init__(self , name , color,raduis):
        super().__init__(name, color) 
        self.raduis =raduis
            
    def  calculate_area(self):
        return round( (pi * self.raduis * self.raduis) , 2)
    
    
class Rectangle(Shape):
    def __init__(self , name , color,length , width):
        super().__init__(name, color) 
        self.length =length
        self.width =width
        
    def  calculate_area(self):
        return round((self.length * self.width),2)
```

```python
square= Square("square1","black" , 5)
rectangle= Rectangle("rectangle1","black" , 5 , 5)
circle= Circle("circle1","black" , 5)

print(f"Area of Square with side length = 5 is : {square.calculate_area()} \n")
print(f"Area of Rectangle with length and width = 5 is : {rectangle.calculate_area()}\n")
print(f"Area of Circle with raduis = 5 is : {circle.calculate_area()}\n")
```

## Question 2 
* Create a Python function that reads a file containing student names and their grades,
* calculates the average grade, and identifies students who have failed.


```python
import pandas as pd

students_grades_df = pd.read_excel('/kaggle/input/students-grades/grades.xlsx')
print(students_grades_df)

average_students_grades = round(students_grades_df['grade'].mean() , 2)
print(f"average_students_grades: {average_students_grades}")

failed_students = students_grades_df[students_grades_df['grade'] < 5]
print(f"Failed Students Names: {failed_students['name'].tolist()}")
```

## Question 3
* given this list : numbers = [4, 9, 16, 25, 36],  
* Calculate square root of each number using lambda and map 
* create dictionary contains numbers and sqrt_of_numbers and write it to json file.

```python
import math
import json

input_list = [4, 9, 16, 25, 36]

# Q1
result_iterator = map(lambda x:math.sqrt(x) , input_list)
result_list =[elem for elem in result_iterator]

print(result_list)

#Q2
dic = {}
for num , sqrt_num in zip(input_list ,result_list ):
    dic[num]=sqrt_num
print(dic)

json_object = json.dumps(dic)
 
with open("sqrt_num.json", "w") as sqrt_num:
    sqrt_num.write(json_object)
```
